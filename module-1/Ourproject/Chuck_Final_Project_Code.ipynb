```python
#assessing size of original data set
df.shape
```

```python
#created a new variable with just the columns needed for analysis
thu = df.iloc[:, 2:6]
thu
```

```python
#iterated over the columns to remove the dollar sign in effort to convert the objects to floats
dfff = thu.applymap(lambda x: x.split('$'))
dfff
```

```python
#new variable with column list for production budget
s = dfff['production_budget']
s.head(25)
```

```python
#new variable with column list for wordwide gross
v = dfff['worldwide_gross']
v
```

```python
#this list needed the comma replaced with a space in order to convert to a float
fv = s.map(lambda x: x[1])
g = fv.map(lambda x: x.strip())
y = g.map(lambda x: x.replace(',', ''))
prod_budget = y.astype(float)
prod_budget
```

```python
#this list needed the comma replaced with a space in order to convert to a float
jh = v.map(lambda x: x[1])
c = jh.map(lambda x: x.strip())
z = c.map(lambda x: x.replace(',', ''))
worldwide_gross1 = z.astype(float)
worldwide_gross1
```

```python
#added the new columns to the existing dataset
thu['prod_budget'] = prod_budget
thu
thu['world_gross'] = worldwide_gross1
thu
```

```python
#now able to subtract the worldwide gross from production budget and created a new variable
roi1 = thu['world_gross'].map(lambda x: x) - thu['prod_budget'].map(lambda y: y)
roi1
```

```python
#new roi variable added to the dataframe
thu['roi'] = roi1
thu
```

```python
#imported another data set so I can start to sort the data by movie name and match it to genre
df6 = pd.read_csv('imdb.title.basics.csv')
df6
```

```python
#dropped unecessary columns
df7 = df6.drop(columns =['tconst', 'original_title','start_year', 'runtime_minutes'])
df7
```

```python
#rename primary_title to movie in order to match names to genres in original data set
df8 = df7.rename(columns = {'primary_title':'movie'})
df8
```

```python
#sort other dataframe by roi to get a top 500 order
wdf = thu.sort_values(by = ['roi'], ascending = False)
wdf
```

```python
#now we will merge the 2 dataframes
result = wdf.merge(df8, on = 'movie')
result
```

```python
#drop more extraneous columns
jt = result.drop(columns = ['production_budget','domestic_gross','worldwide_gross'])
jt
```

```python
#sort values by movie to prepare to drop duplicates
tr = jt.sort_values(by = 'movie')
tr
```

```python
#drop duplicates from movie column
werq = tr.drop_duplicates(subset = 'movie')
werq.head(35)
```

```python
#sort again by ROI to get rank, and show final workable dataset
fds = werq.sort_values(by = "roi", ascending = False)
fds.head(500)
```
